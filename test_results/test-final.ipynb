```python
from graph.workflow import chatbot
```

```python
def ask_chatbot(query: str):
    result = chatbot.invoke({"query": query})
    return result["final_answer"]
```

```python
while True:
    query = input("\nQuestion: ")
    if query.lower() == "exit":
        break
    answer = ask_chatbot(query)
    print("\nQuestion:")
    print(query)
    print("\nAnswer:")
    print(answer)
```
